```python
# Dynamic Task Assignment with Job Queue multi processing version main function
import multiprocessing
import os
from resource_change import *
from common_function import *

def process_sample(sample):
    sample_dir = '../../../../DikeDataset/malware/'
    save_dir = '../evaluation/perturbated_sample/instruction_change/' 
    
    samples = list_files_by_size(sample_dir)
    #print(samples)
    create_directory(save_dir)

    for sample in samples:
        if '.ipynb' in sample or '.pickle' in sample or '.txt' in sample or '.zip' in sample or '.' not in sample:
            continue
        
        print(sample)

        input_filepath = sample_dir+sample
        
        try:
            new_text = disassemble_and_modify(input_filepath, save_dir)

            if new_text is None:
                print(f"[+] Error : failed to make new_text section.")
            else:
                new_text = modify_headers(input_filepath, new_text)
                modify_section(input_filepath, new_text, save_dir)
        except pefile.PEFormatError:
            continue
            
def worker(input_queue):
    while True:
        sample = input_queue.get()
        if sample is None:
            break
        process_sample(sample)
        input_queue.task_done()

def main():
    sample_dir = '../../../../DikeDataset/malware/'
    save_dir = '../evaluation/perturbated_sample/instruction_change/' 
    
    samples = list_files_by_size(sample_dir)
    create_directory(save_dir)

    # 유효한 샘플만 필터링
    samples = [sample for sample in samples if not any(ext in sample for ext in ['.ipynb', '.pickle', '.txt', '.zip','.dll'])]

    # 작업 큐 생성
    input_queue = multiprocessing.JoinableQueue()

    # CPU 코어 수의 절반만 사용하도록 설정
    num_processes = max(1, multiprocessing.cpu_count() // 2)

    # 워커 프로세스 생성 및 시작
    processes = []
    for _ in range(num_processes):
        p = multiprocessing.Process(target=worker, args=(input_queue,))
        p.start()
        processes.append(p)

    # 작업 큐에 작업 추가
    for sample in samples:
        input_queue.put(sample)

    # 모든 작업이 완료되면 None을 넣어 워커 프로세스를 종료시킴
    input_queue.join()
    for _ in range(num_processes):
        input_queue.put(None)

    # 워커 프로세스 종료
    for p in processes:
        p.join()

    print("All tasks are completed.")

if __name__ == '__main__':
    main()
```
